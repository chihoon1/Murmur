```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ranksums
```

```python
# load cases from case features folder
Ent_Case = pd.read_csv('case_features/Window_Entropy_Case.csv', header=None)
Slope_Case = pd.read_csv('case_features/Window_Slope_Case.csv', header=None)
Mf_Case = pd.read_csv('case_features/Window_Mf_Case.csv', header=None)
```

```python
# load controls from control features folder
Ent_Control = pd.read_csv('control_features/Window_Entropy_Control.csv', header=None)
Slope_Control = pd.read_csv('control_features/Window_Slope_Control.csv', header=None)
Mf_Control = pd.read_csv('control_features/Window_Mf_Control.csv', header=None)
```

```python
col_names = ["Entropy ", "Slope ", "ID", "Hurst Exponent ",
    "Left Slope ", "Right Slope ","Left Tangent ",
    "Right Tangent", "Broadness", "Left Tangent Point", "Right Tangent Point"]
# combine Ent and Slope and Mfcc dataframes for each class
Case = pd.concat([ Ent_Case.T, Slope_Case.T, Mf_Case ], axis=1)
Case.columns = col_names
Control = pd.concat([ Ent_Control.T, Slope_Control.T, Mf_Control ], axis=1)
Control.columns = col_names
Case['Case'] = 1
Control['Case'] = 0
```

```python
# combine Case and Control data
data = pd.concat([Case, Control], axis=0)
# suffle data
data = data.sample(frac=1, random_state= 42).reset_index(drop=True);
```

```python
# density plot of Entropy
sns.kdeplot(Case['Entropy '], fill=True, label='Murmur Present', color='red')
sns.kdeplot(Control['Entropy '], fill=True, label='Murmur Absent', color='green')
```

```python
# willcoxon test for Entropy
wilcoxon_stat = ranksums(Case['Entropy '], Control['Entropy '], alternative="two-sided")
p_val =  wilcoxon_stat.pvalue * 2
if p_val > 0.05:
    print(f"no significant difference in distribution")
else:
    print("Difference in distribution. Thus, predicting power exists")
```

```python
# phik correlation matrix
import phik

phik_corr = pd.DataFrame(features_df.phik_matrix())
fig, axs = plt.subplots(1, 1, figsize=(24,12))

cmap = sns.diverging_palette(120, 20, as_cmap=True)

sns.heatmap(phik_corr, square=True, vmin=-1, vmax=1, cmap=cmap, annot=True)
```

Assumed any correlation between a pair of features >= 0.4 to be correlated. 
Left and Right tangent points are correlated.
Hurst exponent and Left or Right tangent points are correlated.
Hurst exponent and Left tanget are correlated.
Left slope and Left tanget or broadness are correlated.
Right slope and right tangent, right tangent point, or braodness are correlated.
left tangent and right tangent point are correlated.
Right tangent and broadness are correlated.
Broadness and right tangent point are correlated.
Left tangent and left or right tangent point are correlated.
Left tangent point and right tangent point are correlated.

Lastly, Broadness and Right tangent won't be used as features as they failed to reject null hypothesis in Wilcoxon rank sum test.

Entropy, Slope, and left tangent point are the best features to be used because they are discrimatory in predicting murmur accroding to the Wilcoxon rank sum test, and they are highly correlated to the class labels, but they are mostly not correlated to each other.
